```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import  MLPRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import StandardScaler

df_new_clean = df.dropna()

X = df_new_clean.drop('Total', axis=1)
y = df_new_clean['Total']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y = y.values.reshape(-1, 1)
y_scaled = scaler.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled,test_size=0.27,random_state=27)

#MLP Model
print('\n For  MLP \n')
model1 = MLPRegressor()
model1.fit(X_train,y_train)
y_pred1 = model1.predict(X_test)

r2Score1 = r2_score(y_test,y_pred1)
MSE1 = mean_squared_error(y_test,y_pred1)

print(f'R_2 Score1:{r2Score1}')
print(f'Mean Square Error1:{MSE1}')

#Extra Tree Regressor
print('\n For Extra Tree Regressor \n')
model2 = ExtraTreesRegressor()
model2.fit(X_train,y_train)
y_pred2 = model2.predict(X_test)

r2Score2 = r2_score(y_test,y_pred2)
MSE2 = mean_squared_error(y_test,y_pred2)

print(f'R_2 Score2:{r2Score2}')
print(f'Mean Square Error2:{MSE2}')

#Random Forest Regressor
print('\n For Random Forest Regressor \n')
model3 = RandomForestRegressor(n_estimators=1000)
model3.fit(X_train,y_train)
y_pred3 = model3.predict(X_test)


r2Score3 = r2_score(y_test,y_pred3)
MSE3 = mean_squared_error(y_test,y_pred3)

print(f'R_2 Score3:{r2Score3}')
print(f'Mean Square Error3:{MSE3}')

#Support Vector Machine

model4 = SVR(kernel='rbf')
print('**\n Running Support Vector Machine Regressor\n**')
model4.fit(X_train,y_train)
y_pred4 = model4.predict(X_test)


r2Score4 = r2_score(y_test,y_pred4)
MSE4 = mean_squared_error(y_test,y_pred4)

print(f'R_2 Score4:{r2Score4}')
print(f'Mean Square Error4:{MSE4}')

#KNN Regressor
print('\n For KNN \n')
model5 = KNeighborsRegressor(n_neighbors=2)
model5.fit(X_train,y_train)
y_pred5 = model5.predict(X_test)


r2Score5 = r2_score(y_test,y_pred5)
MSE5 = mean_squared_error(y_test,y_pred5)

print(f'R_2 Score5:{r2Score5}')
print(f'Mean Square Error5:{MSE5}')


```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.3, random_state=42)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# Define the ComplexNN class
class ComplexNN(nn.Module):
    def __init__(self):
        super(ComplexNN, self).__init__()
        self.layer1 = nn.Linear(32, 256)  
        self.relu = nn.ReLU()
        self.batch_norm1 = nn.BatchNorm1d(256)
        self.layer2 = nn.Linear(256, 128)
        self.relu = nn.ReLU()
        self.layer3 = nn.Linear(128, 64)
        self.relu = nn.ReLU()
        self.layer4 = nn.Linear(64, 32)
        self.relu = nn.ReLU()
        self.layer5 = nn.Linear(32, 16)
        self.relu = nn.ReLU()
        self.out_put_layer = nn.Linear(16, 1)

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.batch_norm1(x)
        x = self.layer2(x)
        x = self.relu(x)
        x = self.layer3(x)
        x = self.relu(x)
        x = self.layer4(x)
        x = self.relu(x)
        x = self.layer5(x)
        x = self.relu(x)
        x = self.out_put_layer(x)
        return x

# Instantiate the network
net = ComplexNN()

# Define the loss function and optimizer
criterion = nn.MSELoss()  
optimizer = optim.Adam(net.parameters(), lr=0.001, weight_decay=0.01)
# Training the model
num_epochs = 100
loss_values = []  # Store loss values

for epoch in range(num_epochs):
    net.train()  # Set the network to training mode
    optimizer.zero_grad()  # Zero the gradients
    outputs = net(X_train_tensor)  # Forward pass
    loss = criterion(outputs, y_train_tensor)  # Compute the loss
    loss.backward()  # Backward pass
    optimizer.step()  # Update the weights

    # Save the loss for plotting the curve later
    loss_values.append(loss.item())

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Plot the loss curve
plt.plot(loss_values)
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error (Loss)')
plt.show()

# Evaluate the model
net.eval()  # Set the network to evaluation mode
with torch.no_grad():  # Disable gradient computation
    train_outputs = net(X_train_tensor)
    test_outputs = net(X_test_tensor)

# Calculate accuracy using R2 Score for regression
train_r2 = r2_score(y_train_tensor, train_outputs)
test_r2 = r2_score(y_test_tensor, test_outputs)

print(f'Training R2 Score: {train_r2*100:.2f}')
print(f'Test R2 Score: {test_r2*100:.2f}')
```
